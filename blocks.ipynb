```python
from ipythonblocks import BlockGrid
```

```python
grid = BlockGrid(15,15)
```

```python
for block in grid:
    if block.row%2 == 0:
        block.rgb = (128,64,32)
    elif block.col%3 == 0:
        block.rgb = (32,64,128)
    else:
        block.rgb = (64,128,32)
grid
```
