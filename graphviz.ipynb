```python
from graphviz import Digraph
```

```python
dg = Digraph(comment="4U+T", format='png')
```

```python
dg.engine = 'circo'
# circo has mindist option to decrease the distance between nodes
dg.graph_attr['mindist']='0.3'
```

```python
for i in range(1, 99):
    dg.node("{}".format(i), height='0.3', width='0.3', fixedsize='True')
```

```python
def f(x, u=2, t=1):
    return t*(x//10) + u*(x%10)
```

```python
f(19,u=1,t=10)
```

```python
for i in range(1, 99):
    dg.edge("{}".format(i), "{}".format(f(i,u=4,t=1)))
```
